```python
# importing required libraries
import xlrd

import xlwt

from xlutils.copy import copy

import numpy as np

import matplotlib.pyplot as plt

from time import sleep

import sys



# this function display the grouped bar chart of student subject marks
def display_grouped_bar_chart_of_all_students_subject_marks():
    
    student = xlrd.open_workbook("Marks.xls")
    student_sheet = student.sheet_by_index(0)
    
    x = np.arange(6)
    y1 = []
    y2 = []
    y3 = []
    y4 = []
    y5 = []
    students = []
    subjects = []
    
    for i in range(student_sheet.ncols):
        
        for j in range(1,student_sheet.nrows):
            
            if i == 0:
                students.append(student_sheet.cell_value(j,i))
                continue
                
            elif i == 1:
                y1.append(student_sheet.cell_value(j,i))
                continue
                
            elif i == 2:
                y2.append(student_sheet.cell_value(j,i))
                continue
                
            elif i == 3:
                y3.append(student_sheet.cell_value(j,i))
                continue
                
            elif i == 4:
                y4.append(student_sheet.cell_value(j,i))
                continue
                
            elif i == 5:
                y5.append(student_sheet.cell_value(j,i))
                continue
                
                
    for i in range(1,student_sheet.ncols-3):
        
        subjects.append(student_sheet.cell_value(0,i))
                
            
    width = 0.15
    plt.bar(x-0.3 , y1 , width , color = "green")
    plt.bar(x-0.15 , y2 , width , color = "indigo")
    plt.bar(x , y3 , width , color = "yellow")
    plt.bar(x+0.15 , y4 , width , color = "red")
    plt.bar(x+0.3 , y5 , width , color = "hotpink")
    plt.xticks(x,students)
    plt.legend(subjects,bbox_to_anchor=(1.2,1))
    plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=2, 
                    top=1, 
                    wspace=1, 
                    hspace=1)
    plt.show()
    
    try:
        
        is_again = int(input("Enter 0 to go back else press any other key to exit : "))
        
        if is_again == 0:
            return main()
        
        else:
            sys.exit()
            
    except:
        
        print("Exitted")


# this function display groped bar chart of all students subject marks with comparision
def display_grouped_bar_chart_of_all_students_subject_marks_with_comparision():
    
    student = xlrd.open_workbook("Marks.xls")
    student_sheet = student.sheet_by_index(0)
    
    x = np.arange(5)
    y1 = []
    y2 = []
    y3 = []
    y4 = []
    y5 = []
    y6 = []
    students = []
    subjects = []
    
    for j in range(1,student_sheet.nrows):
        
        students.append(student_sheet.cell_value(j,0))
        
        for i in range(1,student_sheet.ncols-3):
            
            if j == 1:
                y1.append(student_sheet.cell_value(j,i))
                continue
                
            elif j == 2:
                y2.append(student_sheet.cell_value(j,i))
                continue
                
            elif j == 3:
                y3.append(student_sheet.cell_value(j,i))
                continue
                
            elif j == 4:
                y4.append(student_sheet.cell_value(j,i))
                continue
                
            elif j == 5:
                y5.append(student_sheet.cell_value(j,i))
                continue
                
            elif j == 6:
                y6.append(student_sheet.cell_value(j,i))
                continue
                
                
    for i in range(1,student_sheet.ncols-3):
        
        subjects.append(student_sheet.cell_value(0,i))
                
            
    width = 0.15
    plt.bar(x-0.3 , y1 , width , color = "green")
    plt.bar(x-0.15 , y2 , width , color = "indigo")
    plt.bar(x , y3 , width , color = "yellow")
    plt.bar(x+0.15 , y4 , width , color = "red")
    plt.bar(x+0.3 , y5 , width , color = "hotpink")
    plt.bar(x+0.45 , y6 , width , color = "grey")
    plt.xticks(x,subjects)
    plt.legend(students,bbox_to_anchor=(1.2,1))
    plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=2, 
                    top=1, 
                    wspace=1, 
                    hspace=1)
    plt.show()
    
    try:
        
        is_again = int(input("Enter 0 to go back else press any other key to exit : "))
        
        if is_again == 0:
            return main()
        
        else:
            sys.exit()
            
    except:
        
        print("Exitted")
    
    
    

# this function display pie char of all students total marks
def display_pie_chart_of_all_students_total_marks():
    
    student = xlrd.open_workbook("Marks.xls")
    student_sheet = student.sheet_by_index(0)
    
    students = []
    total = []
    
    for i in range(1,student_sheet.nrows):
        
        students.append(student_sheet.cell_value(i,0))
        total.append(student_sheet.cell_value(i,6))
      
    explodes = []
    for i in total:
        
        if i == max(total):
            explodes.append(0.2)
            
        elif i == min(total):
            explodes.append(0.1)
            
        else:
            explodes.append(0)
            
    plt.title("Pie Chart Of All Students Total Marks")
    plt.pie(total , labels = students, explode = explodes , shadow = True)
    plt.legend(total , bbox_to_anchor=(1.2,1))
    plt.show()
    try:
        
        is_again = int(input("Enter 0 to go back else press any other key to exit : "))
        
        if is_again == 0:
            return main()
        
        else:
            sys.exit()
            
    except:
        
        print("Exitted")
        
    

# this function display pie chart of all students average marks
def display_pie_chart_of_all_students_average_marks():
    
    student = xlrd.open_workbook("Marks.xls")
    student_sheet = student.sheet_by_index(0)
    
    students = []
    avg = []
    
    for i in range(1,student_sheet.nrows):
        
        students.append(student_sheet.cell_value(i,0))
        avg.append(float(student_sheet.cell_value(i,7)))
        
    explodes = []
    
    for i in avg:
        
        if i == max(avg):
            explodes.append(0.2)
            
        elif i == min(avg):
            explodes.append(0.1)
            
        else:
            explodes.append(0)
            
    plt.title("Pie Chart Of All Students Average Marks")
    plt.pie(avg , labels = students, explode = explodes , shadow = True)
    plt.legend(avg,bbox_to_anchor=(1.2,1))
    plt.show()
    
    try:
        
        is_again = int(input("Enter 0 to go back else press any other key to exit : "))
        
        if is_again == 0:
            return main()
        
        else:
            sys.exit()
            
    except:
        
        print("Exitted")
        
    

# this function display pie chart of all students percentage
def display_pie_chart_of_all_students_percentage():
    
    student = xlrd.open_workbook("Marks.xls")
    student_sheet = student.sheet_by_index(0)
    
    students = []
    per = []
    
    for i in range(1,student_sheet.nrows):
        
        students.append(student_sheet.cell_value(i,0))
        per.append(float(student_sheet.cell_value(i,8)[:-1]))
        
    explodes = []
    
    for i in per:
        
        if i == max(per):
            explodes.append(0.2)
            
        elif i == min(per):
            explodes.append(0.1)
            
        else:
            explodes.append(0)
            
    plt.title("Pie Chart Of All Students Percentage")
    plt.pie(per , labels = students, explode = explodes , shadow = True)
    plt.legend(per,bbox_to_anchor=(1.2,1))
    plt.show()
    
    try:
        
        is_again = int(input("Enter 0 to go back else press any other key to exit : "))
        
        if is_again == 0:
            return main()
        
        else:
            sys.exit()
            
    except:
        
        print("Exitted")
        


# this function display bar graph of all students total marks
def display_bar_graph_of_all_students_total_marks():
    
    student = xlrd.open_workbook("Marks.xls")
    student_sheet = student.sheet_by_index(0)
    
    students = []
    total = []
    
    for i in range(1,student_sheet.nrows):
        
        students.append(student_sheet.cell_value(i,0))
        total.append(student_sheet.cell_value(i,6))
        
    colors = ['red','green','blue','yellow','hotpink','grey']
    plt.title("Bar Graph Of All Students Total Marks")
    plt.xlabel("Students")
    plt.ylabel("Total Marks")
    
    for i in range(len(students)):
        
        plt.bar(students[i] , total[i] , width=0.5 , color=colors[i], label=total[i])
    plt.legend(bbox_to_anchor=(1.5,1))
    plt.show()
    
    try:
        
        is_again = int(input("Enter 0 to go back else press any other key to exit : "))
        
        if is_again == 0:
            return main()
        
        else:
            sys.exit()
            
    except:
        
        print("Exitted")
        
    

# this function display bar graph of all students average marks
def display_bar_graph_of_all_students_average_marks():
    
    student = xlrd.open_workbook("Marks.xls")
    student_sheet = student.sheet_by_index(0)
    
    students = []
    avg = []
    
    for i in range(1,student_sheet.nrows):
        
        students.append(student_sheet.cell_value(i,0))
        avg.append(float(student_sheet.cell_value(i,7)))
        
    plt.title("Bar Graph Of All Students Average Marks")
    plt.xlabel("Students")
    plt.ylabel("Average Marks")
    colors = ['red','green','blue','yellow','hotpink','grey']
    
    for i in range(len(students)):
        
        plt.bar(students[i] , avg[i] , width=0.5 , color = colors[i] , label=avg[i])
        
    plt.legend(bbox_to_anchor=(1.5,1))
    plt.show()
    
    try:
        
        is_again = int(input("Enter 0 to go back else press any other key to exit : "))
        
        if is_again == 0:
            return main()
        
        else:
            sys.exit()
            
    except:
        
        print("Exitted")
        
    

# this function display bar graph of all students percentages
def display_bar_graph_of_all_students_percentage():
    
    student = xlrd.open_workbook("Marks.xls")
    student_sheet = student.sheet_by_index(0)
    
    students = []
    per = []
    
    for i in range(1,student_sheet.nrows):
        
        students.append(student_sheet.cell_value(i,0))
        per.append(float(student_sheet.cell_value(i,8)[:-1]))
        
    plt.title("Bar Graph Of All Students Percentage")
    plt.xlabel("Students")
    plt.ylabel("Percentage")
    colors = ['red','green','blue','yellow','hotpink','grey']
    
    for i in range(len(students)):
        
        plt.bar(students[i] , per[i] , width=0.5,color=colors[i],label=per[i])
        
    plt.legend(bbox_to_anchor=(1.5,1))
    plt.show()
    
    try:
        
        is_again = int(input("Enter 0 to go back else press any other key to exit : "))
        
        if is_again == 0:
            return main()
        
        else:
            sys.exit()
            
    except:
        
        print("Exitted")
        

# this function display bar chart of marks with additional details
def display_bar_chart_of_marks_with_additional_details(l,name):
    
    student = xlrd.open_workbook("Marks.xls")
    student_sheet = student.sheet_by_index(0)
    
    index = l.index(name)
    marks = []
    subjects = []
    
    for i in range(1,student_sheet.ncols):
        
        subjects.append(student_sheet.cell_value(0,i))
        
        if i == student_sheet.ncols - 1:
            marks.append(float(student_sheet.cell_value(index+1,i)[:-1]))
            continue
            
        marks.append(float(student_sheet.cell_value(index+1,i)))
        
     
    plt.subplot(2,1,1)
    plt.title("Bar Graph Of Student Marks")
    plt.ylabel("Subjects")
    plt.xlabel("Marks")
    colors=["red",'green','blue','yellow','hotpink']
    plt.barh(subjects[:-3],marks[:-3],height=0.5,color=colors)
    plt.xticks(marks[:-3])
    
    plt.subplot(2,1,2)
    plt.title("Bar Graph of Percentage,total marks and average marks")
    plt.xlabel("Total marks , Percentage , Average Marks")
    plt.ylabel("Marks")
    colors = ['red','green','blue']
    plt.bar(subjects[-3:],marks[-3:],width=0.5,color=colors)
    plt.yticks(marks[-3:])
    plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=2, 
                    top=1, 
                    wspace=1, 
                    hspace=1)
    plt.show()
    
    try:
        
        is_again = int(input("Enter 0 to go back else press any other key to exit : "))
        
        if is_again == 0:
            return print_student_details(l,name)
        
        else:
            sys.exit()
            
    except:
        
        print("Exitted")
        
    

# this function display pie chart of student marks with adittional details
def display_pie_chart_of_marks_with_additional_details(l,name):
    
    student = xlrd.open_workbook("Marks.xls")
    student_sheet = student.sheet_by_index(0)
    
    index = l.index(name)
    marks = []
    subjects = []
    
    for i in range(1,student_sheet.ncols):
        
        subjects.append(student_sheet.cell_value(0,i))
        
        if i == student_sheet.ncols - 1:
            marks.append(float(student_sheet.cell_value(index+1,i)[:-1]))
            continue
            
        marks.append(student_sheet.cell_value(index+1,i))
     
    total = student_sheet.cell_value(index+1,6)
    average = student_sheet.cell_value(index+1,7)
    percentage = student_sheet.cell_value(index+1,8)
    min_marks = min(marks[:-3])
    max_marks = max(marks[:-3])
    explodes = []
    
    for i in marks:
        
        if i == total:
            explodes.append(0.2)
            
        elif i == average or i == percentage:
            explodes.append(0.1)
            
        elif i == max_marks:
            explodes.append(0.05)
            
        elif i == min_marks:
            explodes.append(0.02)
            
        else:
            explodes.append(0)
            
    plt.title("Pie Chart Of Student Marks")
    plt.pie(marks,labels=subjects , explode=explodes , shadow = True)
    plt.legend(marks,bbox_to_anchor = (1.2,1))
    plt.show()
    
    try:
        
        is_again = int(input("Enter 0 to go back else press any other key to exit : "))
        
        if is_again == 0:
            return print_student_details(l,name)
        
        else:
            sys.exit()
            
    except:
        
        print("Exitted")
        

        
# this function display bar chart of student marks
def display_bar_chart_of_marks(l,name):
    
    student = xlrd.open_workbook("Marks.xls")
    student_sheet = student.sheet_by_index(0)
    
    index = l.index(name)
    marks = []
    subjects = []
    colors = ['red','green','blue','yellow','hotpink']
    
    for i in range(1,student_sheet.ncols-3):
        
        subjects.append(student_sheet.cell_value(0,i))
        marks.append(student_sheet.cell_value(index+1,i))
        
    plt.title("Bar Graph Of Student Marks")
    plt.ylabel("Subjects")
    plt.xlabel("Marks")
    plt.barh(subjects,marks,height=0.5 , color=colors)
    plt.xticks(marks)
    plt.subplots_adjust(right=2)
    plt.show()
    
    try:
        
        is_again = int(input("Enter 0 to go back else press any other key to exit : "))
        
        if is_again == 0:
            return print_student_details(l,name)
        
        else:
            sys.exit()
            
    except:
        
        print("Exitted")
        
    

# this function display pie chart of student marks
def display_pie_chart_of_marks(l,name):
    
    student = xlrd.open_workbook("Marks.xls")
    student_sheet = student.sheet_by_index(0)
    
    index = l.index(name)
    marks = []
    subjects = []
    
    for i in range(1,student_sheet.ncols-3):
        
        subjects.append(student_sheet.cell_value(0,i))
        marks.append(student_sheet.cell_value(index+1,i))
     
    min_marks = min(marks)
    max_marks = max(marks)
    explodes = []
    
    for i in marks:
        
        if i == max_marks:
            explodes.append(0.2)
            
        elif i == min_marks:
            explodes.append(0.1)
            
        else:
            explodes.append(0)
            
    plt.title("Pie Chart Of Student Marks")
    plt.pie(marks,labels=subjects , explode=explodes , shadow = True)
    plt.legend(marks,bbox_to_anchor = (1.05,1))
    plt.show()
    
    try:
        
        is_again = int(input("Enter 0 to go back else press any other key to exit : "))
        
        if is_again == 0:
            return print_student_details(l,name)
        
        else:
            sys.exit()
            
    except:
        
        print("Exitted")
        

        
# this function update the excel sheet
def update_all():
    
    student = xlrd.open_workbook("Marks.xls")
    student_sheet = student.sheet_by_index(0)
    
    new_student = copy(student)
    new_student_sheet = new_student.get_sheet(0)
    
    for i in range(1,student_sheet.nrows):
        
        marks = []
        for j in range(1,student_sheet.ncols-3):
            
            marks.append(student_sheet.cell_value(i,j))
            
        new_student_sheet.write(i,6,sum(marks))
        new_student_sheet.write(i,7,"{:.2f}".format(sum(marks)/5))
        new_student_sheet.write(i,8,'{:.2f}'.format(sum(marks)*100/500)+"%")
        
    new_student.save("Marks.xls")


    
# this function prints student details 
def print_student_details_with_additional_details(l,name):
    
    student = xlrd.open_workbook("Marks.xls")
    student_sheet = student.sheet_by_index(0)
    
    index = l.index(name)
    print(name)
    print("Subjects",end='\t\t\t')
    print("Marks")
    marks = []
    
    for i in range(1,6):
        
        print(student_sheet.cell_value(0,i),end=' '*(len('Probability and statitics       ')-len(student_sheet.cell_value(0,i))))
        print(student_sheet.cell_value(index+1,i))
        marks.append(student_sheet.cell_value(index+1,i))
        
    print(student_sheet.cell_value(0,6),end='\t\t\t\t')
    print(sum(marks))
    print(student_sheet.cell_value(0,7),end='\t\t\t\t')
    print("{:.2f}".format(sum(marks)/5))
    print(student_sheet.cell_value(0,8),end='\t\t\t')
    print('{:.2f}'.format(sum(marks)*100/500)+"%")
    
    new_student = copy(student)
    new_student_sheet = new_student.get_sheet(0)
    new_student_sheet.write(index+1,6,sum(marks))
    new_student_sheet.write(index+1,7,"{:.2f}".format(sum(marks)/5))
    new_student_sheet.write(index+1,8,'{:.2f}'.format(sum(marks)*100/500)+"%")
    
    new_student.save("Marks.xls")
    try:
        
        is_again = int(input("Enter 0 to go back else press any other key to exit : "))
        
        if is_again == 0:
            return print_student_details(l,name)
        
        else:
            sys.exit()
            
    except:
        
        print("Exitted")
        

        
# this function prints student details
def print_student_details(l,name):
    
    student = xlrd.open_workbook("Marks.xls")
    student_sheet = student.sheet_by_index(0)
    
    index = l.index(name)
    
    options = {
    1:print_student_details_with_additional_details,
    2:display_pie_chart_of_marks,
    3:display_bar_chart_of_marks,
    4:display_bar_chart_of_marks_with_additional_details,
    5:display_pie_chart_of_marks_with_additional_details}
    
    print(name)
    print("Subjects",end='\t\t\t')
    print("Marks")
    
    for i in range(1,6):
        
        print(student_sheet.cell_value(0,i),end=' '*(len('Probability and statitics       ')-len(student_sheet.cell_value(0,i))))
        print(student_sheet.cell_value(index+1,i))
        
    try:
        
        for i,j in options.items():
            
            print(str(i)+". "+" ".join(j.__name__.split("_")))
        choose = int(input("Choose option from above list else press any other key to return to back : "))
        
        if choose in options:
            options[choose](l,name)
            
        else:
            return print_students()
        
    except:
        
        return print_students()
    

# this function prints students
def print_students():
    student = xlrd.open_workbook('Marks.xls')
    student_sheet = student.sheet_by_index(0)
    l = []
    
    for i in range(1,student_sheet.nrows):
        print(student_sheet.cell_value(i,0))
        l.append(student_sheet.cell_value(i,0))
    name = input("Enter student name of which you want to see the details else press any other key to return to the back menu : ")
    
    if name in l:
        print_student_details(l,name)
    else:
        return main()

# main function
def main():
    
    try:
        
        update_all()
        
        options = {1:print_students,
        2:display_bar_graph_of_all_students_total_marks,
        3:display_bar_graph_of_all_students_average_marks,
        4:display_bar_graph_of_all_students_percentage,
        5:display_pie_chart_of_all_students_total_marks,
        6:display_pie_chart_of_all_students_average_marks,
        7:display_pie_chart_of_all_students_percentage,
        8:display_grouped_bar_chart_of_all_students_subject_marks,
        9:display_grouped_bar_chart_of_all_students_subject_marks_with_comparision}
        
        for i,j in options.items():
            
            print(str(i)+". "+" ".join(j.__name__.split("_")))
            
        choose = int(input("Choose option from above printed list else press any other key to exit : "))
        
        if choose in options:
            options[choose]()
            
        else:
            sys.exit()
            
    except KeyboardInterrupt:
        print("Program Stooped")
        
    except SystemExit:
        print("Exitted")
        
    except:
        print("Exitted")

        
# calling main function
if __name__ == "__main__":
    main()
```
